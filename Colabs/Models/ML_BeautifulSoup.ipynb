```python
from bs4 import BeautifulSoup
import requests
```

```python
page_link = 'https://wwwn.cdc.gov/nchs/nhanes/search/datapage.aspx?Component=Laboratory&CycleBeginYear=2013'
```

```python
page_response = requests.get(page_link, timeout=5)
```

```python
page_content = BeautifulSoup(page_response.content, "html.parser")
```

```python
textContent = []
table = page_content.find(id="GridView1")
n = len(table)
```

```python
table
```

```python
for i in range(1, 50):
    rows = table.find_all("tr")[i].find("td").text
    textContent.append(rows)
```

```python
textContent
```

```python
import csv
with open('new.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(textContent)
```

```python
newText = []
for i in range(1, 50):
    row = []
    for j in range(0, 4):
        cell = table.find_all("tr")[i].find_all("td")[j].text
        row.append(cell.split(','))
    newText.append(row)
```
